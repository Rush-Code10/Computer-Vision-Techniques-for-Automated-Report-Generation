```python

# !pip -q install tensorflow==2.15.0 opencv-python-headless scikit-image matplotlib reportlab numpy==1.26.4 'protobuf<6.0,>=5.26.1'
import os, io, json, math
import numpy as np
import cv2
import matplotlib.pyplot as plt

from dataclasses import dataclass
from typing import Tuple, List, Dict

from skimage import exposure
from skimage.morphology import remove_small_objects, binary_opening, binary_closing, disk, binary_dilation
from skimage.measure import label, regionprops
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4
from reportlab.lib.utils import ImageReader

import tensorflow as tf
from tensorflow.keras import layers, models, losses, optimizers, backend as K

print("TF:", tf.__version__)
```

```python

@dataclass
class CFG:
    img1_path: str = "orlando2010.jpg"
    img2_path: str = "orlando2023.jpg"
    img_size: Tuple[int, int] = (1024, 1024)  # set to None to keep native size
    normalize: bool = True

    # Registration
    do_registration: bool = True
    ecc_warp_mode: int = cv2.MOTION_HOMOGRAPHY
    ecc_iters: int = 1500
    ecc_eps: float = 1e-7

    # Weak labels
    otsu_blur_ksize: int = 5
    top_percentile: float = 96.0  # combine with Otsu (OR)
    min_change_region_pixels: int = 16

    # Patching
    patch_size: int = 128
    patch_stride: int = 64

    # Pair thresholds (auto-tuned if degenerate)
    change_ratio_pos: float = 0.10  # >= changed
    change_ratio_neg: float = 0.02  # <= unchanged

    # Model / training
    base_filters: int = 32
    embedding_dim: int = 64
    batch_size: int = 8
    epochs: int = 25
    lr: float = 5e-4
    seg_loss_weight: float = 1.0
    con_loss_weight: float = 0.2
    dice_smooth: float = 1.0
    alpha_pos_bce: float = 0.7   # class weight for changed pixels

    # Inference
    prob_threshold: float = 0.35

    # Report
    out_dir: str = "outputs"
    report_path: str = "outputs/orlando_change_report.pdf"

CFG = CFG()
os.makedirs(CFG.out_dir, exist_ok=True)
CFG
```

```python

def read_rgb(path: str) -> np.ndarray:
    img = cv2.imread(path, cv2.IMREAD_COLOR)
    if img is None:
        raise FileNotFoundError(f"Image not found: {path}")
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

def resize_if_needed(img: np.ndarray, target_hw):
    if target_hw is None: return img
    th, tw = target_hw
    if img.shape[0] != th or img.shape[1] != tw:
        img = cv2.resize(img, (tw, th), interpolation=cv2.INTER_AREA)
    return img

def ecc_register(img_ref: np.ndarray, img_mov: np.ndarray, mode, iters, eps) -> np.ndarray:
    ref_gray = cv2.cvtColor(img_ref, cv2.COLOR_RGB2GRAY).astype(np.float32) / 255.0
    mov_gray = cv2.cvtColor(img_mov, cv2.COLOR_RGB2GRAY).astype(np.float32) / 255.0
    warp_matrix = np.eye(3, 3, dtype=np.float32) if mode == cv2.MOTION_HOMOGRAPHY else np.eye(2,3, dtype=np.float32)
    criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, iters, eps)
    try:
        cc, warp_matrix = cv2.findTransformECC(ref_gray, mov_gray, warp_matrix, mode, criteria, None, 5)
        if mode == cv2.MOTION_HOMOGRAPHY:
            aligned = cv2.warpPerspective(img_mov, warp_matrix, (img_ref.shape[1], img_ref.shape[0]),
                                          flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)
        else:
            aligned = cv2.warpAffine(img_mov, warp_matrix, (img_ref.shape[1], img_ref.shape[0]),
                                     flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)
        return aligned
    except cv2.error as e:
        print("ECC failed -> using original moving image. Error:", e)
        return img_mov

def normalize_img(img: np.ndarray) -> np.ndarray:
    img = img.astype(np.float32) / 255.0
    for c in range(3):
        img[..., c] = exposure.rescale_intensity(img[..., c], in_range='image', out_range=(0, 1))
    return img

def prepare_images(cfg: CFG):
    im1 = read_rgb(cfg.img1_path)
    im2 = read_rgb(cfg.img2_path)
    im1 = resize_if_needed(im1, cfg.img_size)
    im2 = resize_if_needed(im2, cfg.img_size)

    if cfg.do_registration:
        im2 = ecc_register(im1, im2, cfg.ecc_warp_mode, cfg.ecc_iters, cfg.ecc_eps)

    if cfg.normalize:
        im1 = normalize_img(im1)
        im2 = normalize_img(im2)

    g1 = cv2.cvtColor((im1*255).astype(np.uint8), cv2.COLOR_RGB2GRAY)
    g2 = cv2.cvtColor((im2*255).astype(np.uint8), cv2.COLOR_RGB2GRAY)
    absd = cv2.absdiff(g1, g2)

    # Otsu
    blur = cv2.GaussianBlur(absd, (cfg.otsu_blur_ksize, cfg.otsu_blur_ksize), 0)
    _, m_otsu = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Top-percentile
    thr = np.percentile(absd, cfg.top_percentile)
    m_pct = (absd >= thr).astype(np.uint8) * 255

    # Combine (OR) and clean
    mask = np.maximum(m_otsu, m_pct)
    bin_mask = mask > 0
    bin_mask = binary_opening(bin_mask, disk(2))
    bin_mask = binary_closing(bin_mask, disk(3))
    bin_mask = remove_small_objects(bin_mask, cfg.min_change_region_pixels)
    weak_mask = (bin_mask.astype(np.uint8) * 255)

    # Sanity stats
    raw_ratio = (mask > 0).mean() * 100.0
    clean_ratio = (weak_mask > 0).mean() * 100.0
    print(f"Weak-label coverage — raw:{raw_ratio:.2f}% clean:{clean_ratio:.2f}%")

    return im1, im2, absd, weak_mask

im1, im2, absd, weak_mask = prepare_images(CFG)
print("Prepared:", im1.shape, im2.shape, weak_mask.shape)
```

```python

# Preview + Naïve Baseline
plt.figure(figsize=(14,4))
plt.subplot(1,4,1); plt.imshow(im1); plt.title("2010"); plt.axis("off")
plt.subplot(1,4,2); plt.imshow(im2); plt.title("2023 (aligned)"); plt.axis("off")
plt.subplot(1,4,3); plt.imshow(absd, cmap='gray'); plt.title("Abs Diff"); plt.axis("off")
plt.subplot(1,4,4); plt.imshow(weak_mask, cmap='gray'); plt.title("Weak Mask"); plt.axis("off")
plt.show()

naive_thr = np.percentile(absd, 97.0)
naive_mask = (absd >= naive_thr).astype(np.uint8)
naive_ratio = naive_mask.mean() * 100.0
print(f"Naïve (abs-diff >=97th pct) changed ratio: {naive_ratio:.2f}%")
```

```python

# Patching
def extract_patches(img1, img2, mask, patch, stride):
    H, W, _ = img1.shape
    ps = patch
    p1s, p2s, ms, ratios = [], [], [], []
    for y in range(0, H-ps+1, stride):
        for x in range(0, W-ps+1, stride):
            p1 = img1[y:y+ps, x:x+ps, :]
            p2 = img2[y:y+ps, x:x+ps, :]
            pm = mask[y:y+ps, x:x+ps]
            r = (pm > 0).mean()
            p1s.append(p1); p2s.append(p2); ms.append((pm>0).astype(np.float32)[..., None]); ratios.append(r)
    return np.stack(p1s), np.stack(p2s), np.stack(ms), np.array(ratios, np.float32)

X1, X2, Y, ratios = extract_patches(im1, im2, weak_mask, CFG.patch_size, CFG.patch_stride)
print("Patches:", X1.shape, Y.shape, "avg change ratio:", float(ratios.mean()))
```

```python

# Siamese U-Net
def conv_block(x, f, k=3):
    x = layers.Conv2D(f, k, padding='same')(x)
    x = layers.BatchNormalization()(x)
    return layers.ReLU()(x)

def encoder_shared(inp, base=32):
    x = conv_block(inp, base)
    x1 = conv_block(x, base); p1 = layers.MaxPool2D()(x1)
    x2 = conv_block(p1, base*2); x2 = conv_block(x2, base*2); p2 = layers.MaxPool2D()(x2)
    x3 = conv_block(p2, base*4); x3 = conv_block(x3, base*4); p3 = layers.MaxPool2D()(x3)
    x4 = conv_block(p3, base*8); x4 = conv_block(x4, base*8)
    return x1, x2, x3, x4

def decoder(diff4, skip3, skip2, skip1, base=32):
    d3 = layers.UpSampling2D()(diff4); d3 = layers.Concatenate()([d3, skip3]); d3 = conv_block(d3, base*4); d3 = conv_block(d3, base*4)
    d2 = layers.UpSampling2D()(d3);   d2 = layers.Concatenate()([d2, skip2]); d2 = conv_block(d2, base*2); d2 = conv_block(d2, base*2)
    d1 = layers.UpSampling2D()(d2);   d1 = layers.Concatenate()([d1, skip1]); d1 = conv_block(d1, base);   d1 = conv_block(d1, base)
    out = layers.Conv2D(1, 1, activation='sigmoid')(d1)
    return out

def build_siamese_unet(patch=128, base=32, embedding_dim=64):
    inp1 = layers.Input((patch, patch, 3))
    inp2 = layers.Input((patch, patch, 3))

    shared_inp = layers.Input((patch, patch, 3))
    shared_model = models.Model(shared_inp, encoder_shared(shared_inp, base=base))

    s1a, s2a, s3a, s4a = shared_model(inp1)
    s1b, s2b, s3b, s4b = shared_model(inp2)

    diff4 = layers.Concatenate()([
        layers.Lambda(lambda a: tf.abs(a))(s4a),
        layers.Lambda(lambda a: tf.abs(a))(s4b),
        layers.Lambda(lambda a: tf.abs(a[0]-a[1]))([s4a, s4b])
    ])
    k1 = layers.Lambda(lambda a: tf.abs(a[0]-a[1]))([s1a, s1b])
    k2 = layers.Lambda(lambda a: tf.abs(a[0]-a[1]))([s2a, s2b])
    k3 = layers.Lambda(lambda a: tf.abs(a[0]-a[1]))([s3a, s3b])

    seg_out = decoder(diff4, k3, k2, k1, base=base)

    gp = layers.GlobalAveragePooling2D()(diff4)
    emb = layers.Dense(embedding_dim)(gp)
    return models.Model([inp1, inp2], [seg_out, emb], name="siamese_unet")

model = build_siamese_unet(CFG.patch_size, CFG.base_filters, CFG.embedding_dim)
model.summary()
```

```python

# Losses
def dice_coef(y_true, y_pred, smooth=1.0):
    y_true_f = K.flatten(y_true); y_pred_f = K.flatten(y_pred)
    inter = K.sum(y_true_f * y_pred_f)
    return (2.*inter + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

def weighted_bce(y_true, y_pred, alpha=0.7):
    # alpha weights positives (changed pixels)
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.clip_by_value(y_pred, 1e-7, 1-1e-7)
    loss = - (alpha*y_true*tf.math.log(y_pred) + (1-alpha)*(1-y_true)*tf.math.log(1-y_pred))
    return tf.reduce_mean(loss)

def seg_loss(y_true, y_pred):
    return 0.5*weighted_bce(y_true, y_pred, alpha=CFG.alpha_pos_bce) + 0.5*(1.0 - dice_coef(y_true, y_pred, smooth=CFG.dice_smooth))

def contrastive_loss(y_pair, dists, margin=1.0):
    y_pair = tf.cast(y_pair, tf.float32)
    loss_pos = y_pair * tf.square(dists)
    loss_neg = (1.0 - y_pair) * tf.square(tf.maximum(margin - dists, 0.0))
    return tf.reduce_mean(loss_pos + loss_neg)
```

```python
def make_pairs_safe(X1, X2, Y, ratios, pos_thr, neg_thr):
    # Identify unchanged vs changed patches
    keep = (ratios <= neg_thr) | (ratios >= pos_thr)
    if keep.sum() == 0:
        # Auto-tune: fallback split at median
        med = float(np.median(ratios))
        print(f"[AutoTune] No clear pairs — using median split at {med:.4f}")
        keep = np.ones_like(ratios, dtype=bool)
        sim_labels = (ratios <= med).astype(np.float32)[:, None]
    else:
        sim_labels = (ratios <= neg_thr).astype(np.float32)[:, None]

    # Filter everything consistently
    X1k, X2k, Yk, rk, sl = X1[keep], X2[keep], Y[keep], ratios[keep], sim_labels[keep]

    return X1k, X2k, Yk, rk, sl

X1f, X2f, Yf, rf, sim_lbl = make_pairs_safe(X1, X2, Y, ratios, CFG.change_ratio_pos, CFG.change_ratio_neg)
print("Filtered:", X1f.shape, Yf.shape, sim_lbl.shape, "similar patches")

train_ds = (
    tf.data.Dataset.from_tensor_slices(((X1f, X2f), (Yf, sim_lbl)))
    .shuffle(4096)
    .batch(CFG.batch_size)
    .prefetch(2)
)

```

```python

optimizer = optimizers.Adam(CFG.lr)

@tf.function
def train_step(batch):
    (bx1, bx2), (by, blabel) = batch
    with tf.GradientTape() as tape:
        seg_pred, emb = model([bx1, bx2], training=True)
        l_seg = seg_loss(by, seg_pred)

        # Build contrastive pairs: half same-class (similar), half cross-class (dissimilar) by roll
        emb_a = emb
        emb_b = tf.roll(emb, shift=1, axis=0)
        bl_a  = blabel
        bl_b  = tf.roll(blabel, shift=1, axis=0)
        sim_pair = tf.cast(tf.logical_and(tf.equal(bl_a, 1.0), tf.equal(bl_b, 1.0)), tf.float32)

        dists = tf.norm(emb_a - emb_b, axis=1)
        l_con = contrastive_loss(sim_pair, dists, margin=1.0)

        loss = CFG.seg_loss_weight*l_seg + CFG.con_loss_weight*l_con

    grads = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    return loss, l_seg, l_con, dice_coef(by, seg_pred)

history = {"loss": [], "seg": [], "con": [], "dice": []}
for e in range(CFG.epochs):
    L, Ls, Lc, D = [], [], [], []
    for batch in train_ds:
        loss, lseg, lcon, d = train_step(batch)
        L.append(float(loss)); Ls.append(float(lseg)); Lc.append(float(lcon)); D.append(float(d))
    print(f"Epoch {e+1}/{CFG.epochs} | loss={np.mean(L):.4f} seg={np.mean(Ls):.4f} con={np.mean(Lc):.4f} dice={np.mean(D):.4f}")
    history["loss"].append(np.mean(L)); history["seg"].append(np.mean(Ls)); history["con"].append(np.mean(Lc)); history["dice"].append(np.mean(D))
```

```python

plt.figure(figsize=(8,4))
plt.plot(history["loss"], label="total"); plt.plot(history["seg"], label="seg"); plt.plot(history["con"], label="contrastive")
plt.legend(); plt.title("Training losses"); plt.xlabel("epoch"); plt.ylabel("loss"); plt.show()

plt.figure(figsize=(6,4))
plt.plot(history["dice"], label="dice")
plt.legend(); plt.title("Dice"); plt.xlabel("epoch"); plt.ylabel("dice"); plt.show()
```

```python

# Inference
def predict_full_image(im1, im2, patch, stride):
    H, W, _ = im1.shape
    prob = np.zeros((H, W), dtype=np.float32)
    cnt  = np.zeros((H, W), dtype=np.float32)
    for y in range(0, H - patch + 1, stride):
        for x in range(0, W - patch + 1, stride):
            p1 = im1[y:y+patch, x:x+patch, :][None, ...]
            p2 = im2[y:y+patch, x:x+patch, :][None, ...]
            seg_pred, _ = model.predict([p1, p2], verbose=0)
            seg = seg_pred[0, ..., 0]
            prob[y:y+patch, x:x+patch] += seg
            cnt[y:y+patch, x:x+patch]  += 1.0
    cnt[cnt==0] = 1.0
    return prob / cnt

prob_map = predict_full_image(im1, im2, CFG.patch_size, CFG.patch_stride)
bin_map  = (prob_map >= CFG.prob_threshold).astype(np.uint8)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(im1); plt.title("2010"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(im2); plt.title("2023"); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(prob_map, cmap='viridis'); plt.title("Pred prob"); plt.axis('off')
plt.show()
```

```python

# Post-process & stats
clean = binary_opening(bin_map.astype(bool), disk(2))
clean = binary_closing(clean, disk(3))
clean = binary_dilation(clean, disk(1))
clean = remove_small_objects(clean, CFG.min_change_region_pixels)
clean_mask = clean.astype(np.uint8) * 255

lbl = label(clean)
props = regionprops(lbl)
areas = [p.area for p in props]
areas_sorted_idx = np.argsort(areas)[::-1]
topk = min(5, len(areas_sorted_idx))
top_regions = [props[i] for i in areas_sorted_idx[:topk]]

changed_pixels = int(clean.sum())
total_pixels = int(clean.size)
percent_change = 100.0 * changed_pixels / total_pixels

print(f"Changed pixels: {changed_pixels} / {total_pixels} ({percent_change:.2f}%)")
print("Top blobs:", [int(r.area) for r in top_regions])

# Save artifacts
ovl = (im2.copy()*255).astype(np.uint8)
ovl[clean_mask>0] = (np.clip(ovl[clean_mask>0].astype(np.int32) + np.array([60,0,0]), 0, 255)).astype(np.uint8)
cv2.imwrite(os.path.join(CFG.out_dir, "orlando2010_preview.png"), cv2.cvtColor((im1*255).astype(np.uint8), cv2.COLOR_RGB2BGR))
cv2.imwrite(os.path.join(CFG.out_dir, "orlando2023_preview.png"), cv2.cvtColor((im2*255).astype(np.uint8), cv2.COLOR_RGB2BGR))
cv2.imwrite(os.path.join(CFG.out_dir, "weak_mask.png"), weak_mask)
cv2.imwrite(os.path.join(CFG.out_dir, "pred_prob.png"), (prob_map*255).astype(np.uint8))
cv2.imwrite(os.path.join(CFG.out_dir, "pred_mask.png"), clean_mask)
cv2.imwrite(os.path.join(CFG.out_dir, "overlay_2023_changes.png"), cv2.cvtColor(ovl, cv2.COLOR_RGB2BGR))

with open(os.path.join(CFG.out_dir, "summary.json"), "w") as f:
    json.dump({
        "changed_pixels": changed_pixels,
        "total_pixels": total_pixels,
        "percent_change": percent_change,
        "top_regions_pixel_area": [int(r.area) for r in top_regions]
    }, f, indent=2)

percent_change
```

```python

# PDF Report
def add_image(c, img_path, x, y, w=None, h=None):
    try:
        img = ImageReader(img_path)
        iw, ih = img.getSize()
        if w is None and h is None:
            w, h = iw, ih
        elif w is None:
            w = iw * (h/ih)
        elif h is None:
            h = ih * (w/iw)
        c.drawImage(img, x, y, width=w, height=h)
    except Exception as e:
        c.drawString(50, y+20, f"Could not load image: {img_path} ({e})")

def generate_pdf(cfg: CFG, stats: Dict):
    c = canvas.Canvas(cfg.report_path, pagesize=A4)
    W, H = A4

    c.setFont("Helvetica-Bold", 16)
    c.drawString(50, H-50, "Orlando Infrastructure Change Report (2010 → 2023) — Patched")

    c.setFont("Helvetica", 10)
    c.drawString(50, H-70, "Method: Siamese CNN + contrastive; Weak labels: Otsu ∪ Top-Percentile; Registration: ECC Homography.")

    img_w = (W - 80) / 3
    img_h = img_w * 0.75
    y1 = H - 110 - img_h
    add_image(c, os.path.join(cfg.out_dir, "orlando2010_preview.png"), 30, y1, w=img_w, h=img_h)
    add_image(c, os.path.join(cfg.out_dir, "orlando2023_preview.png"), 40+img_w, y1, w=img_w, h=img_h)
    add_image(c, os.path.join(cfg.out_dir, "weak_mask.png"), 50+2*img_w, y1, w=img_w, h=img_h)
    c.setFont("Helvetica", 9)
    c.drawString(30, y1-12, "2010 image")
    c.drawString(40+img_w, y1-12, "2023 image (aligned)")
    c.drawString(50+2*img_w, y1-12, "Weak change mask")

    y2 = y1 - 40 - img_h
    add_image(c, os.path.join(cfg.out_dir, "pred_prob.png"), 30, y2, w=img_w, h=img_h)
    add_image(c, os.path.join(cfg.out_dir, "pred_mask.png"), 40+img_w, y2, w=img_w, h=img_h)
    add_image(c, os.path.join(cfg.out_dir, "overlay_2023_changes.png"), 50+2*img_w, y2, w=img_w, h=img_h)
    c.drawString(30, y2-12, "Predicted change (probability)")
    c.drawString(40+img_w, y2-12, "Predicted change (mask)")
    c.drawString(50+2*img_w, y2-12, "2023 with changes overlay")

    with open(os.path.join(cfg.out_dir, "summary.json")) as f:
        stats = json.load(f)

    y_stats = y2 - 40
    c.setFont("Helvetica-Bold", 12); c.drawString(30, y_stats, "Summary Stats")
    y_stats -= 16
    c.setFont("Helvetica", 10)
    c.drawString(30, y_stats, f"Changed pixels: {stats['changed_pixels']:,} / {stats['total_pixels']:,} ({stats['percent_change']:.2f}%)")
    y_stats -= 14
    if stats['top_regions_pixel_area']:
        c.drawString(30, y_stats, "Top change blobs (pixel areas): " + ", ".join(map(str, stats['top_regions_pixel_area'])))

    c.setFont("Helvetica-Oblique", 9)
    c.drawString(30, 30, "Note: Convert pixels to area using GSD (m/px).")

    c.showPage(); c.save()
    return cfg.report_path

pdf_path = generate_pdf(CFG, {})
pdf_path
```
